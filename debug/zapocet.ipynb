```python
import numpy as np
from matplotlib import pyplot as plt
import cv2

def rgb2gray(img,mez):
    """
    Autor: Zajan Ond≈ôej
    Vytvo≈ôeno: 5.1.21
    img - p≈Øvodn√≠ ƒçernob√≠l√Ω obr√°zek
    mez - p≈ôi p≈ôekroƒçen√≠, pixel je vyhodnocen jako ƒçern√Ω <1,255>
    return - bin√°rn√≠ obr√°zek - matice
    """
    mez=mez*3
    binary_img=np.zeros(np.shape(img)[0:2])
    for row_idx,row in enumerate(img):
        for col_idx, pix in enumerate(row):
            if((int(pix[0])+int(pix[1])+int(pix[2]))>mez):
                binary_img[row_idx,col_idx]=1#jedniƒçka je zde jako ƒçern√°
    return binary_img
img_orig = plt.imread("obr3.jpg")
#print(len(img))
#print(np.shape(img))
#plt.imshow(img)

#p≈ôevod na ƒçernob√≠l√Ω obr√°zek
img=rgb2gray(img_orig,97)
print("Velikost obr√°ztku= {}".format(np.shape(img)))
img=np.logical_not(img).astype(int)
plt.imshow(img,cmap='Greys')
```
